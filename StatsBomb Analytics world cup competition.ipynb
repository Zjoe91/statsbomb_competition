```python
import pandas as pd
```

```python
#read the file

df = pd.read_excel('wc_event_data.xlsx')
```

```python
# show data

df.head()
```

```python
df.columns
```

```python
# show all players where they have center forward position under the position_name column
new_df = df[df['position_name'] == 'Center Forward']

new_df.head()

        
```

```python
# show all players in the new_df created before where they have clearance, block, duel and interception 

new_df = new_df[new_df['type_name'].isin(['Clearance', 'Block', 'Duel', 'Interception'])]

new_df["player_name"]

```

```python
# show top players with the highest number of events

new_df["player_name"].value_counts().head(10)
```

```python
new1_df = df[df["pass_body_part_name"].isin(['Right Foot', 'Left Foot'])]
#for every player name find the number of passes done with the left foot and with the right foot. If the player has has made more passes with one foot than the other we just keep the one with the least passes completed.
new1_df.groupby(["player_name", "pass_body_part_name"]).size()
#rank the players based on the number of passes they have made with the least foot. From that number of passes with that foot, rank the players based on the most number of passes made
new1_df.groupby(["player_name", "pass_body_part_name"]).size().unstack().fillna(0).min(axis=1).sort_values(ascending=False)
```

```python
# order new1_df by pass_body_part_name frequency in descending order, find the first player that appears twice

new2_df = new1_df.groupby(["player_name", "pass_body_part_name"]).size().sort_values(ascending=False)
new2_df
```
